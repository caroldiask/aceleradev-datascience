## modelos

```python
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
```

```python
regressores= [
    SVR(),
    KNeighborsRegressor(n_jobs=-1),
    DecisionTreeRegressor(random_state=13),
    RandomForestRegressor(random_state=13, n_jobs=-1),
    MLPRegressor(random_state=13)]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import math
```

```python
X = df_treino.drop(columns= 'NU_NOTA_MT').values
y = df_treino['NU_NOTA_MT'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=13)
```

```python
for regr in regressores:
    
    # rodada de treino:
    regr.fit(X_train, y_train)
    
    # rodada de teste:
    y_predito = regr.predict(X_test)
    
    # pegando o nome do algoritmo
    regressor = regr.__class__.__name__
    
    # imprimindo os resultados
    print("="*33)
    print(regressor)
    print("=== Resultado ===")
    print("Mean Squared Error (MSE)", mean_squared_error(y_test, y_predito))
    print("Root Mean Squared Error (RMSE)", math.sqrt(mean_squared_error(y_test, y_predito)))
    print("Mean Absolute Error (MAE)", mean_absolute_error(y_test, y_predito))
    print("R Squared (R²)", r2_score(y_test, y_predito))
```

```python
RFR = RandomForestRegressor(random_state=13, n_jobs=-1)
```

```python
RFR.fit(X_train, y_train)
y_pred= RFR.predict(X_test)
```
